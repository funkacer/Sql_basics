```python
tables_df = pd.read_sql_query(sql='''SELECT *
FROM pg_catalog.pg_tables
WHERE schemaname != 'pg_catalog' AND 
    schemaname != 'information_schema';
''', con=conn)
```

```python
tables = tables_df['tablename'].to_list()
```

```python
tables
```

```python
for table in tables:
    table_df = pd.read_sql_query(sql='''
    SELECT * FROM {};
    '''.format(table), con=conn)
    print(table)
    table_df.to_sql(name='{}'.format(table), if_exists='replace', con=connl)
```
