```python
pd.read_sql_query(sql = '''
SELECT * FROM film;
''', con = conn)
```

```python
pd.read_sql_query(sql = '''
SELECT * FROM film f
LEFt JOIN film_category fc
ON f.film_id = fc.film_id
LEFt JOIN category c
ON fc.category_id = c.category_id
LEFT JOIN film_actor fa
ON f.film_id = fa.film_id 
LEFT JOIN actor a
ON fa.actor_id = a.actor_id
ORDER BY f.film_id;
''', con = conn)
```

```python
# some film have no actor

pd.read_sql_query(sql = '''
SELECT * FROM category c
JOIN film_category fc
ON fc.category_id = c.category_id
JOIN film f
ON fc.film_id = f.film_id
JOIN film_actor fa
ON fa.film_id = f.film_id
JOIN actor a
ON a.actor_id = fa.actor_id
ORDER BY f.film_id;
''', con = conn)
```

```python
# some film have no info

pd.read_sql_query(sql = '''
WITH t1 AS (
SELECT * FROM film f
LEFt JOIN film_category fc
ON f.film_id = fc.film_id
LEFt JOIN category c
ON fc.category_id = c.category_id
LEFT JOIN film_actor fa
ON f.film_id = fa.film_id 
LEFT JOIN actor a
ON fa.actor_id = a.actor_id
ORDER BY f.film_id)

SELECT DISTINCT *
FROM t1
WHERE actor_id IS NULL
;
''', con = conn)
```

```python
pd.read_sql_query(sql = '''
SELECT * FROM film f
LEFt JOIN film_category fc
ON f.film_id = fc.film_id
LEFt JOIN category c
ON fc.category_id = c.category_id
LEFT JOIN inventory i
ON f.film_id = i.film_id;
''', con = conn)
```

```python
pd.read_sql_query(sql = '''
SELECT DISTINCT film_id, store_id, COUNT(*) AS film_in_store_count FROM inventory
GROUP BY 1, 2
ORDER BY 1, 2;
''', con = conn)
```

```python
pd.read_sql_query(sql = '''
SELECT * FROM inventory
WHERE film_id = 1;
''', con = conn)
```

```python
# all film_id from inventory have film record
pd.read_sql_query(sql = '''
SELECT DISTINCT i.film_id, f.title
FROM inventory i
LEFT JOIN film f
ON i.film_id = f.film_id
WHERE f.title IS NULL;
''', con = conn)
```

```python
# all film_id from inventory have film record, 3 have no actor
pd.read_sql_query(sql = '''
SELECT DISTINCT i.film_id, f.title
FROM inventory i
LEFT JOIN film f
ON i.film_id = f.film_id
LEFT JOIN film_actor fa
ON f.film_id = fa.film_id
WHERE fa.actor_id IS NULL;
''', con = conn)
```
