```python
def init_db(db_name):
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()
    cursor.execute("CREATE TABLE '20200118_condition' ("+
            "_id INTEGER PRIMARY KEY,"+
            "city TEXT,"+
            "country TEXT,"+
            "region TEXT,"+
            "chill INTEGER,"+
            "direction INTEGER,"+
            "speed REAL,"+
            "sunrise TEXT,"+
            "sunset TEXT,"+
            "humidity INTEGER,"+
            "pressure REAL,"+
            "rising INTEGER,"+
            "visibility REAL,"+
            "pubdate TEXT,"+
            "text TEXT,"+
            "code INTEGER,"+
            "temperature INTEGER,"+
            "creatat TEXT,"+
            "changeat TEXT)")
    
    #cursor.execute("DROP TABLE forecast")
    cursor.execute("CREATE TABLE '20200118_forecast' ("+
            "_id INTEGER PRIMARY KEY,"+
            "condate TEXT,"+
            "day TEXT,"+
            "date TEXT,"+
            "code INTEGER,"+
            "low INTEGER,"+
            "high INTEGER,"+
            "text TEXT,"+
            "creatat TEXT,"+
            "changeat TEXT)")
        
    cursor.close()
```

```python
init_db('Weather_db.db')
```
