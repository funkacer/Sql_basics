```python
import sqlite3
conn = sqlite3.connect('Weather_db_20200118.db')
cursor = conn.cursor()
cursor.execute('''SELECT "Total" as "text", "Total" as "pubdate", CAST(COUNT(*)* 100.0 / sum(count(*)) over() AS DECIMAL(5,2)) AS "Record_Count"

FROM "condition"

WHERE "creatat" >= "2020-01-01" and "creatat" < "2020-01-02"

UNION ALL

SELECT "text", "pubdate", CAST(COUNT(*)* 100.0 / sum(count(*)) over() AS DECIMAL(5,2)) AS "Record_Count"

FROM "condition"

WHERE "creatat" >= "2020-01-01" and "creatat" < "2020-01-02"

GROUP BY "pubdate"

ORDER BY 2 ASC

 ''')
data = cursor.fetchall()
cursor.close()
```
