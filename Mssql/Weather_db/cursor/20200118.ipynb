```python
def init_db(sch_name):
    conn_string = "Driver={SQL Server};Server=DESKTOP-NO9U58U\SQLEXPRESS;Database=dbMoje;Trusted_Connection=yes;"
    conn = pyodbc.connect(conn_string)
    cursor = conn.cursor()

    try:
        cursor.execute('CREATE SCHEMA [{}]'.format(sch_name))
        conn.commit()
    except Exception as e:
        print(e)

    try:
        cursor.execute("CREATE TABLE [{}].[20200118_condition] (".format(sch_name)+
                "_id INTEGER PRIMARY KEY,"+
                "city TEXT,"+
                "country TEXT,"+
                "region TEXT,"+
                "chill INTEGER,"+
                "direction INTEGER,"+
                "speed REAL,"+
                "sunrise TEXT,"+
                "sunset TEXT,"+
                "humidity INTEGER,"+
                "pressure REAL,"+
                "rising INTEGER,"+
                "visibility REAL,"+
                "pubdate TEXT,"+
                "text TEXT,"+
                "code INTEGER,"+
                "temperature INTEGER,"+
                "creatat TEXT,"+
                "changeat TEXT)")
        conn.commit()
    except Exception as e:
        print(e)

    try:
        #cursor.execute("DROP TABLE forecast")
        cursor.execute("CREATE TABLE [{}].[20200118_forecast] (".format(sch_name)+
                "_id INTEGER PRIMARY KEY,"+
                "condate TEXT,"+
                "day TEXT,"+
                "date TEXT,"+
                "code INTEGER,"+
                "low INTEGER,"+
                "high INTEGER,"+
                "text TEXT,"+
                "creatat TEXT,"+
                "changeat TEXT)")
        conn.commit()
    except Exception as e:
        print(e)

    cursor.close()
```

```python
init_db(schema_name)
```
