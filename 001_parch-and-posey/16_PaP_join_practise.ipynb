SELECT r.name region, a.name account, 
       o.total_amt_usd/(o.total + 0.01) unit_price
FROM region r
JOIN sales_reps s
ON s.region_id = r.id
JOIN accounts a
ON a.sales_rep_id = s.id
JOIN orders o
ON o.account_id = a.id;

```python
pd.read_sql_query(sql='''
SELECT r.name region, a.name account, o.total_amt_usd/(o.total+0.01) unit_price
FROM sales_reps s
JOIN region r
ON s.region_id = r.id
JOIN accounts a
ON a.sales_rep_id = s.id
JOIN orders o
ON o.account_id = a.id;
''', con=conn)
```

```python
pd.read_sql_query(sql='''
SELECT r.name region, a.name account, o.id order_id, total_amt_usd / (o.total + 0.01) unit_price
FROM sales_reps s
JOIN region r
ON s.region_id = r.id
JOIN accounts a
ON s.id = a.sales_rep_id
JOIN orders o
ON o.account_id = a.id
ORDER BY o.id;
''', con=conn)
```

```python
# check against number of orders, e.g. assume 1 order = 1 account = 1 sales rep = 1 region
pd.read_sql_query(sql='''
SELECT *, total_amt_usd / (o.total + 0.01) unit_price
FROM orders o;
''', con=conn)
```
