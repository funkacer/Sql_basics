```python
import sqlite3
import pandas as pd
from IPython.display import display
from ipywidgets import HTML
```

```python
conn =sqlite3.connect('parch-and-posey.db')
```

```python
tables = pd.read_sql_query(sql = 'select * from sqlite_master where type = "table";', con = conn)
```

# NOT IN

```python
pd.read_sql_query(sql = '''
SELECT name, primary_poc, sales_rep_id
FROM accounts
WHERE name NOT IN ('Walmart', 'Target', 'Nordstrom');
''', con = conn)
```

```python
pd.read_sql_query(sql = '''
SELECT *
FROM web_events
WHERE channel NOT IN ('organic', 'adwords');
''', con = conn)
```

```python
display(HTML(pd.read_sql_query(sql = '''
SELECT *
FROM web_events 
WHERE account_id NOT IN (1001, 1011);
''', con = conn).to_html()))
```

# NOT LIKE

```python
pd.read_sql_query(sql = '''
SELECT *
FROM accounts
WHERE name NOT LIKE ('C%');
''', con = conn)
```

```python
pd.read_sql_query(sql = '''
SELECT *
FROM accounts
WHERE name NOT LIKE ('%one%');
''', con = conn)
```

```python
pd.read_sql_query(sql = '''
SELECT *
FROM accounts
WHERE name NOT LIKE ('%s')
''', con = conn)
```
