Task 1:

Use the accounts table to create first and last name columns that hold the first and last names for the primary_poc.

```python
# sqlite version

pd.read_sql_query(sql='''
SELECT SUBSTR(primary_poc, 1, INSTR(primary_poc, ' ')-1) first_name,
SUBSTR(primary_poc, INSTR(primary_poc, ' ') +1, LENGTH (primary_poc) - INSTR(primary_poc, ' ')) last_name
FROM accounts;
''', con=conn)
```

Solution:

SELECT LEFT(primary_poc, STRPOS(primary_poc, ' ') -1 ) first_name, 
RIGHT(primary_poc, LENGTH(primary_poc) - STRPOS(primary_poc, ' ')) last_name
FROM accounts;

Task 2:

Now see if you can do the same thing for every rep name in the sales_reps table. Again provide first and last name columns.

```python
# sqlite version

pd.read_sql_query(sql='''
SELECT SUBSTR(name, 1, INSTR(name, ' ')-1) first_name,
SUBSTR(name, INSTR(name, ' ') +1, LENGTH (name) - INSTR(name, ' ')) last_name
FROM sales_reps;
''', con=conn)
```

Solution:

FROM sales_reps;
