Task 4:

Provide the name for each region for every order, as well as the account name and the unit price they paid (total_amt_usd/total) for the order. However, you should only provide the results if the standard order quantity exceeds 100. Your final table should have 3 columns: region name, account name, and unit price. In order to avoid a division by zero error, adding .01 to the denominator here is helpful total_amt_usd/(total+0.01).

```python
pd.read_sql_query(sql='''
SELECT r.name r_name, a.name a_name, o.total_amt_usd/(o.total+0.01) o_unit_price
FROM sales_reps s
JOIN region r
ON s.region_id = r.id
JOIN accounts a
ON s.id = a.sales_rep_id
JOIN orders o
ON a.id = o.account_id
AND o.standard_qty > 100;
''', con=conn)
```

```python
#Solution:
pd.read_sql_query(sql='''
SELECT r.name region, a.name account, o.total_amt_usd/(o.total + 0.01) unit_price
FROM region r
JOIN sales_reps s
ON s.region_id = r.id
JOIN accounts a
ON a.sales_rep_id = s.id
JOIN orders o
ON o.account_id = a.id
WHERE o.standard_qty > 100;
''', con=conn)
```

Task 7:

What are the different channels used by account id 1001? Your final table should have only 2 columns: account name and the different channels. You can try SELECT DISTINCT to narrow down the results to only the unique values.

```python
pd.read_sql_query(sql='''
SELECT DISTINCT a.name a_name, w.channel
FROM accounts a
JOIN web_events w
ON a.id = w.account_id
AND a.id = 1001
''', con=conn)
```

```python
#POKUS
pd.read_sql_query(sql='''
SELECT a.name a_name, w.channel, COUNT(*) count
FROM accounts a
JOIN web_events w
ON a.id = w.account_id
AND a.id = 1001
GROUP BY w.channel
''', con=conn)
```

```python
# Solution:
pd.read_sql_query(sql='''
SELECT DISTINCT a.name, w.channel
FROM accounts a
JOIN web_events w
ON a.id = w.account_id
WHERE a.id = '1001';
''', con=conn)
```

Task 8:

Find all the orders that occurred in 2015. Your final table should have 4 columns: occurred_at, account name, order total, and order total_amt_usd.

```python
pd.read_sql_query(sql='''
SELECT DISTINCT o.occurred_at, a.name, o.total, o.total, total_amt_usd
FROM accounts a
JOIN orders o
ON a.id = o.account_id
AND o.occurred_at BETWEEN '2015-01-01' AND '2016-01-01'
ORDER BY o.occurred_at;
''', con=conn)
```

```python
# Solution:
pd.read_sql_query(sql='''
SELECT o.occurred_at, a.name, o.total, o.total_amt_usd
FROM accounts a
JOIN orders o
ON o.account_id = a.id
WHERE o.occurred_at BETWEEN '2015-01-01' AND '2016-01-01'
ORDER BY o.occurred_at DESC;
''', con=conn)
```
