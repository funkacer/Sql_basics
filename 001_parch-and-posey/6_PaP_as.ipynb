# AS

```python
pd.read_sql_query(sql = '''
SELECT id, (standard_amt_usd/total_amt_usd)*100 AS std_percent, total_amt_usd
FROM orders
LIMIT 10;
''', con = conn)
```

```python
pd.read_sql_query(sql = '''
SELECT id, account_id, standard_amt_usd / standard_qty AS std_unit_price
FROM orders
ORDER BY std_unit_price
''', con = conn)
```

```python
# POZOR NA INT
pd.read_sql_query(sql = '''
SELECT id, account_id, standard_amt_usd, standard_qty, standard_amt_usd/standard_qty AS std_unit_price
FROM orders
WHERE standard_qty != 0
ORDER BY std_unit_price
''', con = conn)
```

```python
# POZOR NA INT
pd.read_sql_query(sql = '''
SELECT id, account_id, standard_amt_usd, standard_qty, standard_amt_usd*1.0/standard_qty AS std_unit_price
FROM orders
WHERE standard_qty != 0
ORDER BY std_unit_price
''', con = conn)
```

```python
pd.read_sql_query(sql = '''
SELECT id, account_id, poster_amt_usd, total_amt_usd,
poster_amt_usd/(standard_amt_usd+gloss_amt_usd +poster_amt_usd) AS poster_usd_perc
FROM orders
limit 10
''', con = conn)
```
