```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('parch-and-posey.db')
```

```python
cursor = conn.cursor()
cursor.execute('''
select * from sqlite_master where type = "table";
''')
columns = [col[0] for col in cursor.description]
data = cursor.fetchall()
cursor.close()
```

```python
pd.DataFrame(data, columns=columns)
```

```python
pd.read_sql_query(sql='''
SELECT orders.*
FROM orders;
''', con=conn)
```

```python
pd.read_sql_query(sql='''
SELECT orders.*
FROM orders
JOIN accounts
ON orders.account_id = accounts.id;
''', con=conn)
```

```python
pd.read_sql_query(sql='''
SELECT orders.*, accounts.*
FROM orders
JOIN accounts
ON orders.account_id = accounts.id;
''', con=conn)
```

```python
pd.read_sql_query(sql='''
SELECT *
FROM orders
JOIN accounts
ON orders.account_id = accounts.id;
''', con=conn)
```
