```python
import sqlite3
import pandas as pd
```

```python
conn =sqlite3.connect('parch-and-posey.db')
```

```python
tables = pd.read_sql_query(sql = 'select * from sqlite_master where type = "table";', con = conn)
```

# With numeric variables

```python
pd.read_sql_query(sql = '''
SELECT *
FROM orders
WHERE gloss_amt_usd >= 1000
LIMIT 5;
''', con = conn)
```

```python
pd.read_sql_query(sql = '''
SELECT *
FROM orders
WHERE total_amt_usd < 500
LIMIT 10;
''', con = conn)
```

# With nonnumeric variables

```python
pd.read_sql_query(sql = '''
SELECT name, website, primary_poc
FROM accounts
WHERE name = 'Exxon Mobil';
''', con = conn)
```

```python
pd.read_sql_query(sql = '''
SELECT name, website, primary_poc
FROM accounts
WHERE name != 'Exxon Mobil';
''', con = conn)
```
