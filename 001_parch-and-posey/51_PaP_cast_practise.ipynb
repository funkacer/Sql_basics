Task 4:

Write a query to change the date to the correct SQL date format. You will need to use at least SUBSTR and CONCAT to perform this operation.

```python
# sqlite version SELECT CAST(b AS INT) + 2 AS alias FROM A

pd.read_sql_query(sql='''
WITH t1 AS 
(SELECT [index] indx, SUBSTR(date,1,INSTR(date,'/')-1) month,
SUBSTR(SUBSTR(date, INSTR(date,'/')+1,LENGTH(date) - INSTR(date,'/')), 1, INSTR(SUBSTR(date, INSTR(date,'/')+1,LENGTH(date) - INSTR(date,'/')), '/')-1) day,
SUBSTR(SUBSTR(SUBSTR(date, INSTR(date,'/')+1,LENGTH(date) - INSTR(date,'/')), INSTR(SUBSTR(date, INSTR(date,'/')+1,LENGTH(date) - INSTR(date,'/')), '/')+1, LENGTH(date)-INSTR(SUBSTR(date, INSTR(date,'/')+1,LENGTH(date) - INSTR(date,'/')), '/')), 1, INSTR(SUBSTR(SUBSTR(date, INSTR(date,'/')+1,LENGTH(date) - INSTR(date,'/')), INSTR(SUBSTR(date, INSTR(date,'/')+1,LENGTH(date) - INSTR(date,'/')), '/')+1, LENGTH(date)-INSTR(SUBSTR(date, INSTR(date,'/')+1,LENGTH(date) - INSTR(date,'/')), '/')), ' ')-1) year 

FROM sf_crime_data)

SELECT s.*, t1.year || '-' || t1.month || '-' || t1.day cdate
FROM t1
JOIN sf_crime_data s
ON t1.indx = s.[index];
''', con=conn)
```

Solution:

SELECT date orig_date, (SUBSTR(date, 7, 4) || '-' || LEFT(date, 2) || '-' || SUBSTR(date, 4, 2)) new_date

Task 5:

Once you have createt a column in the correct format, use either CAST or :: to convert this to date.

```python
# sqlite version SELECT CAST(b AS INT) + 2 AS alias FROM A

pd.read_sql_query(sql='''
WITH t1 AS 
(SELECT [index] indx, SUBSTR(date,1,INSTR(date,'/')-1) month,
SUBSTR(SUBSTR(date, INSTR(date,'/')+1,LENGTH(date) - INSTR(date,'/')), 1, INSTR(SUBSTR(date, INSTR(date,'/')+1,LENGTH(date) - INSTR(date,'/')), '/')-1) day,
SUBSTR(SUBSTR(SUBSTR(date, INSTR(date,'/')+1,LENGTH(date) - INSTR(date,'/')), INSTR(SUBSTR(date, INSTR(date,'/')+1,LENGTH(date) - INSTR(date,'/')), '/')+1, LENGTH(date)-INSTR(SUBSTR(date, INSTR(date,'/')+1,LENGTH(date) - INSTR(date,'/')), '/')), 1, INSTR(SUBSTR(SUBSTR(date, INSTR(date,'/')+1,LENGTH(date) - INSTR(date,'/')), INSTR(SUBSTR(date, INSTR(date,'/')+1,LENGTH(date) - INSTR(date,'/')), '/')+1, LENGTH(date)-INSTR(SUBSTR(date, INSTR(date,'/')+1,LENGTH(date) - INSTR(date,'/')), '/')), ' ')-1) year 

FROM sf_crime_data)

SELECT s.*, CAST((t1.year || '-' || t1.month || '-' || t1.day) AS DATE) cdate
FROM t1
JOIN sf_crime_data s
ON t1.indx = s.[index];
''', con=conn)
```

Solution:

SELECT date orig_date, (SUBSTR(date, 7, 4) || '-' || LEFT(date, 2) || '-' || SUBSTR(date, 4, 2))::DATE new_date
FROM sf_crime_data;
