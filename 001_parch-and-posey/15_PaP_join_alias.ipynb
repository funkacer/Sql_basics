# Alias is often 1-letter, AS can be omitted

but cant combine alias with name:

SELECT w.id, a.id
FROM web_events w
JOIN accounts a
ON web_events.account_id = accounts.id

```python
pd.read_sql_query(sql='''
SELECT w.id, a.id
FROM web_events AS w
JOIN accounts AS a
ON w.account_id = a.id
''', con=conn)
```

```python
pd.read_sql_query(sql='''
SELECT w.id AS web_id, a.id AS account_id
FROM web_events AS w
JOIN accounts AS a
ON w.account_id = a.id
''', con=conn)
```

```python
pd.read_sql_query(sql='''
SELECT w.id web_id, a.id account_id
FROM web_events w
JOIN accounts a
ON w.account_id = a.id
''', con=conn)
```
